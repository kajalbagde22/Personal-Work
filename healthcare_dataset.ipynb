Groupby / Crosstab / Pivot_table

Basic Group By:

Group the data by gender and calculate the mean age for each gender.


```python
df.groupby('gender')['age'].mean().reset_index()
```

Group the data by work_type and calculate the count of individuals in each work type.

```python
df.groupby('work_type').size().reset_index()
```

Multiple Aggregations:

Group the data by Residence_type and calculate the mean and standard deviation of avg_glucose_level and bmi for each residence type.


```python
a = df.groupby('Residence_type').agg({'avg_glucose_level': ['mean', 'std'],'bmi': ['mean', 'std']})
a
```

Group the data by smoking_status and stroke and calculate the average age and bmi.


```python
b = df.groupby(['smoking_status', 'stroke']).agg({ 'age': 'mean','bmi': 'mean'}).reset_index()
b
```

Conditional Group By:

Group the data by hypertension and calculate the proportion of individuals who have had a stroke.


```python
df.groupby('hypertension')['stroke'].mean().reset_index()
```

Group the data by heart_disease and calculate the mean age for individuals with and without strokes.


```python
df.groupby(['heart_disease', 'stroke'])['age'].mean().reset_index()
```

Custom Aggregations:

Group the data by work_type and aggregate the following:
Mean avg_glucose_level
Maximum age
Count of stroke occurrences


```python
c = df.groupby('work_type').agg(
    mean_avg_glucose_level=('avg_glucose_level', 'mean'),
    max_age=('age', 'max'),
    count_stroke=('stroke', 'sum')
).reset_index()
```

Pivot Table Questions
Basic Pivot Table:

Create a pivot table showing the mean avg_glucose_level for each combination of gender 
and Residence_type.

```python
a = pd.pivot_table(data=df,index='gender',columns='Residence_type',values='avg_glucose_level',aggfunc='mean',margins=True)
a
```

Create a pivot table showing the count of strokes for each work_type, segmented by 
hypertension.


```python
b = pd.pivot_table(data=df,index='work_type',columns='hypertension',values='stroke',aggfunc='count',margins=True)
b
```

Multiple Aggregations in Pivot Table:

Create a pivot table that shows the mean and median age for each combination of 
smoking_status and ever_married.


```python
c = pd.pivot_table(df,values='age',index=['smoking_status','ever_married'],aggfunc=['mean','median'],margins=True)
c
```

Create a pivot table with work_type and gender as the index, and calculate the mean, 
minimum, and maximum bmi.

```python
d = pd.pivot_table(df,values='bmi',index=['work_type','gender'],aggfunc=['mean','min','max'],margins=True)
d
```

Pivot Table with Custom Functions:

Create a pivot table that shows the percentage of strokes for each combination of work_type 
and Residence_type. Use a custom lambda function to calculate the percentage

```python
e = pd.pivot_table(df,index=['work_type','Residence_type'],values='stroke',aggfunc=lambda x: (x.sum ()/ len(x)) * 100)
e
```
